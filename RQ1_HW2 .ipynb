Caricando il primo file per il mese di gennaio dell'anno 2018, ci accorgiamo che il nostro dataset contiene dei dati che non servono alla nostra analisi come per esempio delle corse che iniziano in mesi diversi da gennaio ed anni diversi dal 2018 ma sono registrate sul file di gennaio 2018. Dal momento che il numero di questi dati è esiguo (esempio 342 casi per gennaio 2018) rispetto al totale, decidiamo di trascurali dalla nostra analisi e quindi di eliminarli dal nostro dataset. Tale rimozione non andrebbe ad influenzare la nostra analisi finale. 2) Consideriamo le corse a cavallo della mezzanotte come corse appartenenti al giorno in cui sono partite. 3) Consideriamo il distretto di appartenza quello di partenza.

creiamo una variabile booleana all'interno della quale mettiamo True o False a seconda se l'anno e il mese coincidono con 2018-01 oppure no. Dopodichè lo trasformo in una serie panda e lo applico per filtrare il mio dataset. Otteniamo, così, un dataset ripulito con solo le corse di Gennaio 2018.

```python
pickup_datetime = []
dropoff_datetime = []
res = []
for date in df_j['tpep_pickup_datetime']:
    if('2018-01' in date):
        pickup_datetime.append(True)
    else: 
        pickup_datetime.append(False)
        
for date in df_j['tpep_dropoff_datetime']:
    if(('2018-01' in date) or ('2018-02-01' in date)): 
        dropoff_datetime.append(True)
    else:
        dropoff_datetime.append(False)

for i in range(len(pickup_datetime)):
    if (pickup_datetime[i] == False) or (dropoff_datetime[i]==False):
        res.append(False)
    else:
        res.append(True)
        
is_january = pd.Series(res) 
new_df_j = df_j[is_january]
new_df_j
```

Usiamo le espressioni regolari per eliminare l'orario da tpep_pickup_datetime per poter fare successivamente il groupby su una copia pulita del dataset

```python
temp_df_j = df_join.replace({
    'tpep_pickup_datetime': '[0-9]{2}:[0-9]{2}:[0-9]{2}'
}, '', regex=True)
temp_df_j
```

Raggruppo il mio dataset per data e distretto e faccio un conteggio delle corse totali giornaliere per il mese di Gennaio nei vari distretti. 

```python
g = temp_df_j.groupby(['tpep_pickup_datetime','Borough']).tpep_pickup_datetime.count()
g
```

Creo un dizionario che ha per chiave i vari distretti e per valori il totale delle corse per il mese di Gennaio. 

```python
d = g.to_dict()
d1 = {}
for i in d:
    if i[1] not in d1:
        d1[i[1]] = d[i]
    else: 
        d1[i[1]] += d[i]
d1
  
```

Ci vogliamo quindi calcolare la media giornaliera di corse per ogni distretto. 

Dal momento che all'interno dei nostri distretti ne compare uno sotto la voce di 'Unknown' abbiamo pensato di ridistribuire tali valori, equamente all'ineterno dei vari distretti al fine di non perdere i 158904 dati. 

```python
if 'Unknown' in d1:
    for i in d1:
        d1[i] += round(((d1[i]/(rows-d1['Unknown']))*100)*(d1['Unknown']/100))
    d1.pop(i)
d1
```

```python
daily_races = {}
for i in d1:
    daily_races[i] = round(d1[i]/len(g.keys().levels[0]),2)
daily_races 
```

```python
df_daily_races = pd.DataFrame.from_dict(daily_races, orient = 'index', columns=['January'])
```
